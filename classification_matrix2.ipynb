```python
import pandas as pd
import numpy as np
import numpy as np
import pandas as pd
df=pd.read_csv('heart.csv')
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
clf1=LogisticRegression()
clf2=DecisionTreeClassifier()
```

```python
clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)
```

```python
y_prd1=clf1.predict(x_test)
y_prd2=clf2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print(accuracy_score(y_test,y_prd1))
print(accuracy_score(y_test,y_prd2))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prd1))
print(classification_report(y_test,y_prd2))
```

```python
confusion_matrix(y_test,y_prd1)
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
print(f1_score(y_test,y_prd1))
print(precision_score(y_test,y_prd1))
print(recall_score(y_test,y_prd1))
```

```python
print(f1_score(y_test,y_prd2))
print(precision_score(y_test,y_prd2))
print(recall_score(y_test,y_prd2))
```

```python
precision_score(y_test,y_prd2,average=None)
```

```python
recall_score(y_test,y_prd2,average=None)
```

```python
#multi class classification=two or more classes
from sklearn.metrics import classification_report
print(classification_report(y_test,y_prd1))
print(classification_report(y_test,y_prd2))
```
