```python
import pandas as pd

# Load the Titanic dataset

titanic_df = pd.read_csv('tested.csv')

# View initial columns
print("Columns:", titanic_df.columns)

# Define the custom transformation function for normalizing fare within each group
def normalize_fare(group):
    group['Normalized_Fare'] = (group['Fare'] - group['Fare'].mean()) / group['Fare'].std()
    return group

# Apply the function to each group in 'Pclass'
titanic_df = titanic_df.groupby('Pclass').apply(normalize_fare)

# Display the first few rows to verify the transformation
titanic_df[['Pclass', 'Fare', 'Normalized_Fare']].head()

```
