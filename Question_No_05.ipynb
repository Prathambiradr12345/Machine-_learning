```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
titanic_df = pd.read_csv('tested.csv')

# Set up the figure and primary axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Fare on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Passenger Index')
ax1.set_ylabel('Fare', color=color)
ax1.plot(titanic_df.index, titanic_df['Fare'], color=color, label='Fare')
ax1.tick_params(axis='y', labelcolor=color)

# Set up the secondary y-axis
ax2 = ax1.twinx()
color = 'tab:green'
ax2.set_ylabel('Age', color=color)
ax2.plot(titanic_df.index, titanic_df['Age'], color=color, label='Age')
ax2.tick_params(axis='y', labelcolor=color)

# Add a title and show the plot
plt.title('Fare and Age of Passengers on the Titanic')
fig.tight_layout()  #  layout to prevent overlap
plt.show()
```
