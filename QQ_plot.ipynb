```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df= sns.load_dataset('iris')
```

```python
sns.kdeplot(df['sepal_length'])
```

```python
temp=sorted(df['sepal_length'].tolist())
```

```python
y_quant=[]
for i in range(1,101):
  y_quant.append(np.percentile(temp,i))
```

```python
samples=np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant=[]
for i in range(1,101):
  x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

```python
#using Statismodel
import statsmodels.api as sm
import matplotlib.pyplot as plt
fig=sm.qqplot(df['sepal_length'],line='45',fit=True)
plt.title('QQ plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```

```python
#uniform
samples=np.random.uniform(low=0,high=1,size=1000)
plt.hist(samples,bins=100)
plt.show()
```

```python
#pareto
import numpy as np
import matplotlib.pyplot as plt
alpha=3
xm=1
x=np.linspace(0.1,10,100)
y=alpha * (xm**alpha) / (x**(alpha+1))
plt.plot(x,y)
plt.plot(np.log(x),np.log(y))
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('pareto distribution')
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm  # Keep this import

alpha = 2  # Changed variable name to alpha
sm_scale = 1  # Renamed variable to avoid conflict
x = stats.pareto.rvs(b=alpha, scale=sm_scale, size=1000)

plt.hist(x)

pareto = stats.pareto.fit(x, floc=0)
dist = stats.pareto(b=pareto[0], scale=pareto[2])

fig = sm.qqplot(x, dist=dist, line='45')  # Now using the correct sm (statsmodels.api)
plt.title('QQ plot')
plt.xlabel('Theoretical Quantiles')
plt.ylabel('Sample Quantiles')
plt.show()
```
