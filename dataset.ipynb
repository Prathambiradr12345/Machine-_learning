```python
#family type
def trans_family_size(family_size):
    if family_size == 1:
        return 'alone'
    elif family_size == 2 and family_size <= 5:
      return 'Small'
    else:
      return'large'
```

```python
df['Family_type']=df['family_Size'].apply(trans_family_size)
```
