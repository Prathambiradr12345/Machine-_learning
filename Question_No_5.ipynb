```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Set hierarchical indexing using 'Pclass' and 'Sex'
titanic_hierarchical = titanic_data.set_index(['Pclass', 'Sex'])

# Display the hierarchical DataFrame
print("Hierarchical DataFrame:")
print(titanic_hierarchical.head())

# Access data for first class passengers
first_class = titanic_hierarchical.loc[1]
print("\nFirst Class Passengers:")
print(first_class)

# Access data for female passengers across all classes
female_passengers = titanic_hierarchical.loc[(slice(None), 'female'), :]
print("\nFemale Passengers:")
print(female_passengers)

# Group by 'Pclass' and 'Sex' and calculate the sum of survivors
survivors_by_class_and_sex = titanic_hierarchical.groupby(level=['Pclass', 'Sex'])['Survived'].sum()
print("\nSurvivors by Class and Sex:")
print(survivors_by_class_and_sex)

# Reset the hierarchical index
titanic_reset = titanic_hierarchical.reset_index()
print("\nDataFrame after resetting index:")
print(titanic_reset.head())
```
