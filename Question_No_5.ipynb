```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Calculate survival rate and average age by passenger class
summary = titanic_data.groupby('Pclass').agg(Survival_Rate=('Survived', 'mean'), Average_Age=('Age', 'mean')).reset_index()

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the survival rate on the first y-axis
ax1.bar(summary['Pclass'], summary['Survival_Rate'] * 100, color='b', alpha=0.6, label='Survival Rate (%)')
ax1.set_xlabel('Passenger Class')
ax1.set_ylabel('Survival Rate (%)', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Create a secondary y-axis for average age
ax2 = ax1.twinx()
ax2.plot(summary['Pclass'], summary['Average_Age'], color='r', marker='o', label='Average Age')
ax2.set_ylabel('Average Age', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Set the title and show the plot
plt.title('Survival Rate and Average Age by Passenger Class')
fig.tight_layout()
plt.show()
```
