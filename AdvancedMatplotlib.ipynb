```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
batters=pd.read_csv('batter.csv')
batters
```

```python
batters.info()
```

```python
sm=batters.head(100).sample(25,random_state=5)
```

```python
sm
```

```python
plt.figure(figsize=(10,10))
plt.scatter(sm['avg'],sm['strike_rate'],s=sm['runs'])
for i in range(sm.shape[0]):
  plt.text(sm['avg'].iloc[i],sm['strike_rate'].iloc[i],sm['batter'].iloc[i])
```

```python
plt.figure(figsize=(10,10))
plt.scatter(sm['avg'],sm['strike_rate'],s=sm['runs'])
plt.axhline(130,color='red')
plt.axvline(30,color='red')
for i in range(sm.shape[0]):
  plt.text(sm['avg'].iloc[i],sm['strike_rate'].iloc[i],sm['batter'].iloc[i])
```

```python
fig,ax=plt.subplots(figsize=(10,10))
ax.scatter(batters['avg'],batters['strike_rate'],color='grey')
ax.set_xlabel('avg')
ax.set_ylabel('strike_rate')
ax.set_title('avg vs strike_rate')
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,figsize=(7,7))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='orange')
ax[0].set_xlabel('avg')
ax[0].set_ylabel('strike_rate')
ax[0].set_title('avg vs strike_rate')
ax[1].scatter(batters['avg'],batters['runs'],color='red')
ax[1].set_xlabel('avg')
ax[1].set_ylabel('runs')
ax[1].set_title('avg vs runs')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0, 0].scatter(batters['avg'], batters['strike_rate'], color='orange')  # Access the subplot using [row, col]
ax[0, 0].set_xlabel('avg')
ax[0, 0].set_ylabel('strike_rate')
ax[0, 0].set_title('avg vs strike_rate')

ax[0, 1].scatter(batters['avg'], batters['runs'], color='red')  # Access the second subplot in the first row
ax[0, 1].set_xlabel('avg')
ax[0, 1].set_ylabel('runs')
ax[0, 1].set_title('avg vs runs')



plt.show()
```
