```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
ax[0, 0].scatter(batters['avg'], batters['strike_rate'], color='orange')  # Access the subplot using [row, col]
ax[0, 0].set_xlabel('avg')
ax[0, 0].set_ylabel('strike_rate')
ax[0, 0].set_title('avg vs strike_rate')

ax[0, 1].scatter(batters['avg'], batters['runs'], color='red')  # Access the second subplot in the first row
ax[0, 1].set_xlabel('avg')
ax[0, 1].set_ylabel('runs')
ax[0, 1].set_title('avg vs runs')



plt.show()
```

3D surface plots

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
```

```python
x
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
z=xx**2+yy**2
```

```python
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=fig.add_subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

contour plots

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='jet')
fig.colorbar(p)
```

```python
delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery
```

```python
delivery.head()
```

```python
delivery['ballnumber'].unique()
```

```python
temp=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid=temp.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(grid)
plt.colorbar()
```
