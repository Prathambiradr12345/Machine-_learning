```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n=10#no of trails
p=0.5  #probabiltity of success
size=1000  # no of samples
b=np.random.binomial(n,p,size)
plt.hist(b,density=True)
plt.show()
```

```python
no=10000
sz=300
dr=(0,1)
samples=np.random.uniform(dr[0],dr[1],(no,sz))
samples_mean=np.mean(samples,axis=1)
plt.hist(samples_mean,density=True,bins=30,edgecolor='black')
plt.title("histogram means")
plt.xlabel('sample mean')
plt.ylabel('density')
plt.show()
```
