```python
batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='black',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dotted')
plt.title('comparison')
plt.xlabel('year')
plt.ylabel('runs')
plt.legend(['V Kohli','RG Sharma'])
plt.show()
```

```python
bt=pd.read_csv('batsman_season_record.csv')
```

```python
bt
```

```python
plt.bar(np.arange(bt.shape[0])-0.2,bt['2017'],width=0.2)
plt.bar(np.arange(bt.shape[0]),bt['2016'],width=0.2)
plt.bar(np.arange(bt.shape[0])+0.2,bt['2015'],width=0.2)
plt.xticks(np.arange(bt.shape[0]),bt['batsman'])
plt.legend(['2015','2016','2017'])
plt.show()
```

```python
#
plt.bar(bt['batsman'],bt['2015'])
plt.bar(bt['batsman'],bt['2016'],bottom=bt['2015'])
plt.bar(bt['batsman'],bt['2017'],bottom=(bt['2016']+bt['2015']))

plt.legend(['2015','2016','2017'])
plt.show()
```
