```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Resample the data to 10-minute intervals, calculating the mean only for numeric columns
resampled_data = titanic_data.resample('10min').agg(
    {col: 'mean' for col in titanic_data.select_dtypes(include=['number']).columns}
)
# Explanation:
# 1. We use .agg() for more control over aggregation.
# 2. We create a dictionary to specify 'mean' for numeric columns only.
# 3. titanic_data.select_dtypes(include=['number']).columns selects only the numeric columns.

# Print the resampled data
resampled_data.head()
```

# **Rolling Statitics**

```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Calculate rolling mean and standard deviation
rolling_mean = titanic_data['Age'].rolling(window=10).mean()
rolling_std = titanic_data['Age'].rolling(window=10).std()

# Print the rolling statistics
print(rolling_mean.head())
print(rolling_std.head())
```

# **Laging**

```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Calculate lagged values
lagged_age = titanic_data['Age'].shift(1)
lagged_survived = titanic_data['Survived'].shift(1)

# Print the lagged values
print(lagged_age.head())
print(lagged_survived.head())
```

# **Exponetial Smoothing**

```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Calculate exponential smoothing
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
ses_model = SimpleExpSmoothing(titanic_data['Age'])
ses_model_fit = ses_model.fit()

# Print the smoothed values
print(ses_model_fit.fittedvalues.head())
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Handle missing values in 'Age' column before decomposition
titanic_data['Age'].fillna(titanic_data['Age'].mean(), inplace=True)

# Decompose the time series after handling missing values
# Adjust the period to match the data's actual seasonality (or frequency)
# For example, if you expect hourly seasonality, use period=60 (60 minutes)
# or you could determine your frequency by autocorrelation and partialautocorrelation
# functions
decomposition = sm.tsa.seasonal_decompose(titanic_data['Age'], model='additive', period=60)

# Print the decomposed components
print(decomposition.trend.head())
print(decomposition.seasonal.head())
print(decomposition.resid.head())
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the Titanic dataset
titanic_data = pd.read_csv('tested.csv')

# Create a time series index
titanic_data['Boarding Time'] = pd.date_range(start='1912-04-10', periods=len(titanic_data), freq='min')
titanic_data.set_index('Boarding Time', inplace=True)

# Fit an ARIMA model
from statsmodels.tsa.arima.model import ARIMA
arima_model = ARIMA(titanic_data['Age'], order=(1,1,1))
arima_model_fit = arima_model.fit()

# Print the ARIMA model summary
df=arima_model_fit.summary()
df
```
